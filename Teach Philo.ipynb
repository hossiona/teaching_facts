#### I beleive that teaching is much more than just delivering knowledge in a [convenient format][1] to the students. The teacher’s first tasks to **motivate** their student’s interest and ensure the students are excited about the subject matter. Students are exposed to problems where they must find information and data broadly rather than narrowly from their own intellectual capabilities in a cooperative environment.One of my most important functions is helping students learn: how to [learn effectively][2], how to [effectively solve problems][3], and how to judge the long-range impact of their solutions **(active learning)**. My active learning of Classroom Assessment Techniques (CATs) with state-of-the-art **pedagogical approach** imply several active learning of techniques of metacognition using self-rating, engagement, critical thinking, problem solving, writing synopsis of important topics, motivation through group students and in-class writing. These activities help students improve their metacognitive skill, increase their interest and heighten their engagement.

I always desire to get the comments & questions from students. #### I was working as an IRACDA (Institutional Research and Academic Career Development Award) postdoctoral research fellow at the University of Kansas (KU) and KU-Haskell teaching fellow (Lecturer) at Haskell Indian Nations University Fall 2014 to Fall 2016. As an Assistant Professor of Chemistry full time, I taught several Chemistry I & II, University chemistry II, Essentials of Organic chemistry in the Department of Chemistry at Fort Hays State University in Fall 2016 and Spring 2017. Then, from Fall 2017 to Spring 2021, I taught full time both lectures and laboratories for General Chemistry GOB (Principle of General Organic and Biological Chemistry), General Chemistry I & II and/or Chemistry I & II and Organic chemistry I & II in Department of Chemistry (Sciences) at Cloud County Community College. From Fall 2021 to current, I am teaching full time Chemistry courses including Introductory Chemistry, General Chemistry I and II, Intro Biochemistry, Organic Chemistry I and II, Princ. Organic Chemistry, Analytical Chemistry, and Princ. biochemistry.  I rely on all of the available tools and methods that are available in order to teach my students the core concepts of courses. The students in my classrooms arrive with a **diverse set of learning needs** and a range of cultural experiences and identities. Learning, and the ways that a person’s brain changes in response to new experiences, must be looked at within the context and the culture in which they occur. I was able to implement the unique blend of face-to-face and virtual course works to assist student learning due to COVID-19 pandemic in Spring 2020 and Fall 2020. I have developed the Principle of General Organic and Biological Chemistry asynchronous DL both lecture and hands-on lab activity in online. Moreover, I have developed an [Intro Chemistry lab manual][10], [General Chemistry I lab manual][11], and [Organic Chemistry lab manual][12] to teach in a convenient format.
